**4. The NLP pipeline**

```python
raw = open('NLP.txt').read()
```

```python
type(raw)
```

```python
tokens = word_tokenize(raw)
type(tokens)
```

```python
words = [w.lower() for w in tokens]
type(words)
```

```python
vocab = sorted(set(words))
type(vocab)
```
